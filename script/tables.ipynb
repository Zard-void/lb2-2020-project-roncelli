```python
import pandas as pd
from scipy.stats import sem
```

```python
grid_results = pd.read_csv('../models/results.tsv', sep='\t', index_col=0)
grid_results = grid_results.drop(['mean_fit_time', 'std_fit_time', 'mean_score_time', 'std_score_time', 'param_random_state', 'params', 'rank_test_score', 'std_train_score', 'std_test_score'], axis=1)

grid_results_multi = pd.MultiIndex.from_frame(grid_results[['param_kernel', 'param_C', 'param_gamma']], names=['Kernel', 'C', 'gamma'])
grid_results.index = grid_results_multi
grid_results.drop(['param_C','param_gamma', 'param_kernel'], axis=1, inplace=True)
```

```python
grid_results_test = grid_results.iloc[:, 6:]
grid_results_test= grid_results_test.sort_index()
grid_results_test.columns = [1,2,3,4,5,'Average']
grid_results_test = grid_results_test.drop(('linear', 4, 2))
grid_results_test = grid_results_test.drop(('linear', 2, 2))
grid_results_test['Ser'] = grid_results_test.iloc[:, 0:5 ].sem(axis=1)
grid_results_test
```

```python
grid_results_test.to_latex(float_format="\\num{%f}", buf='../grid_results_train.txt', multirow=True, escape=False)
```

```python
grid_results_test = grid_results.iloc[:, :6]
grid_results_test= grid_results_test.sort_index()
grid_results_test.columns = [1,2,3,4,5,'Average']
grid_results_test = grid_results_test.drop(('linear', 4, 2))
grid_results_test = grid_results_test.drop(('linear', 2, 2))
grid_results_test['Ser'] = grid_results_test.iloc[:, 0:5 ].sem(axis=1)
grid_results_test.to_latex(float_format="\\num{%f}", buf='../grid_results_test.txt', multirow=True, escape=False)
grid_results_test
```
